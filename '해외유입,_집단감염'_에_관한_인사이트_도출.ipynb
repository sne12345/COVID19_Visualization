## 2-4 색깔 지정 (해외유입 - darkblue, 접촉 감염 -  orange, etc - red)

```python
df_map['color']=0
for x in range(len(df_map)):
    if (df_map['infection_case'][x] == 'overseas inflow'):    
        df_map['color'][x] = 'blue'
    else :    
        df_map['color'][x] = 'orange'
#     else:
#         df_map['color'][x] = 'red'   # 미확인
df_map
```

## 2-5 서울 지역의 해외 유입 및 환자 접촉 감염을 표시 
### - contact with patient = 주황색, overseas inflow = 남색

```python
m = folium.Map(location=[37.5600,126.98], tiles="cartodbpositron", zoom_start=12)
for x in range(len(df_map)):
    location = [df_map['latitude'][x],df_map['longitude'][x]]
    radius=df_map['count'][x]*3.4

    folium.CircleMarker(
        location,
        radius,
        popup=df_map['infection_case'][x],
        color=df_map['color'][x],
        fill=True,
        fill_color=df_map['color'][x],
        control_scale=True
        
    ).add_to(m)
m
```

## 3-2 해외 유입 확진자 중 연령대별 비율 구하기

### - 파이차트 함수 만들기

```python
def pie_chart(data, col, title = ''):
    cnt_df = data[col].value_counts().reset_index()
    fig = px.pie(cnt_df, 
                 values = col, 
                 names = 'index', 
                 title = title, 
                 template = 'seaborn',
                 color_discrete_sequence=px.colors.sequential.RdBu)
    fig.update_traces(rotation=90, pull=0.05, textinfo="value+percent+label")
    fig.show()
```
